```python
eng_marks = {}
eng_marks.update({'Harsh': 99, 'Aadhi': 50, 'Karunesh': 15, 'Sam': 19})
print("eng_marks =", eng_marks)
```

```python
lang_marks = {'Harsh': 100, 'Aadhi': 50, 'Karunesh': 90, 'Sam': 50}
print("lang_marks =", lang_marks)
```

```python
actual_marks = []
actual_marks.append(eng_marks)
actual_marks.append(lang_marks)
print("actual_marks =", actual_marks)
```

```python
totalMarks = []
for i in range(len(actual_marks[0])):  # Assuming both dictionaries have same keys
    total = actual_marks[0][list(actual_marks[0].keys())[i]] + actual_marks[1][list(actual_marks[1].keys())[i]]
    totalMarks.append(total)
print("totalMarks =", totalMarks)
```

```python
totalMarks = []
for eng_mark, lang_mark in zip(eng_marks.values(), lang_marks.values()):
    total = eng_mark + lang_mark
    totalMarks.append(total)
print("totalMarks =", totalMarks)
```

```python
subjects = ['Eng', 'Lang']
list_of_name_sub_tot = [(name, subjects[0], subjects[1], 'total') for name in eng_marks.keys()]
print("list_of_name_sub_tot =", list_of_name_sub_tot)
```

```python
class_marks = {name_tuple: [eng_marks[name_tuple[0]], lang_marks[name_tuple[0]], sum(marks)] for name_tuple, marks in zip(list_of_name_sub_tot, zip(eng_marks.values(), lang_marks.values()))}
print("class_marks =", class_marks)
```

```python
for name_tuple, marks in class_marks.items():
    total = marks[2]
    if total > 120:
        marks.append("Kutty Pattas")
    elif total > 100:
        marks.append("Good Student")
    else:
        marks.append("Slow bloomer")
print("class_marks =", class_marks)
```
