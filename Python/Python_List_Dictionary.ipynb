```python
def cube_elements(lst):
    return [x**3 for x in lst]

lst = [3, 6, 9, 12, 15]
lst = cube_elements(lst)
print(lst)  # Output: [27, 216, 729, 1728, 3375]
```

```python
def min_value_in_dict(sample_dict):
    return min(sample_dict.values())

sample_dict = {'Physics': 82, 'Math': 65, 'history': 75}
min_val = min_value_in_dict(sample_dict)
print(min_val)  # Output: 65
```

```python
class Animal:
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        print("Woof")

class Cat(Animal):
    def sound(self):
        print("Meow")

my_dog = Dog()
my_cat = Cat()

my_dog.sound()  # Output: Woof
my_cat.sound()  # Output: Meow
```
