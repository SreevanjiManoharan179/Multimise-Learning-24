```python
import math
import sys
```

```python
radius = 5  # Example radius
area = math.pi * radius ** 2
print(f"Area of the circle: {area}")
```

```python
number = 6789.987
ceil_value = math.ceil(number)
floor_value = math.floor(number)
print(f"Ceil value: {ceil_value}")
print(f"Floor value: {floor_value}")
```

```python
number = 8943
factorial_value = math.factorial(number)
sys.set_int_max_str_digits(0)  # Increase the maximum string length

print(f"Factorial of {number}: {factorial_value}")
```

```python
num1 = 25
num2 = 120
gcd_value = math.gcd(num1, num2)
print(f"GCD of {num1} and {num2}: {gcd_value}")
```

```python
number = -9780.67
absolute_value = math.fabs(number)
print(f"Absolute value of {number}: {absolute_value}")
```

```python
deposit = 30000
interest_rate = 0.04  # 4%
years = 5

final_amount = deposit * pow(1 + interest_rate, years)
print(f"Final amount deposited: {final_amount}")
```

```python
side1 = 6
side2 = 8
hypotenuse = math.sqrt(side1 ** 2 + side2 ** 2)
print(f"Hypotenuse: {hypotenuse}")
```

```python
number = 10
log_base_e = math.log(number)
log_base_3 = log_base_e / math.log(3)
print(f"Natural logarithm of {number} with base 3: {log_base_3}")
```

```python
sin_value = math.sin(10)
cos_value = math.cos(0)
tan_value = math.tan(1)
print(f"Sin(10): {sin_value}")
print(f"Cos(0): {cos_value}")
print(f"Tan(1): {tan_value}")
```

```python
degrees1 = 180
degrees2 = 100
radians1 = math.radians(degrees1)
radians2 = math.radians(degrees2)
print(f"{degrees1} degrees in radians: {radians1}")
print(f"{degrees2} degrees in radians: {radians2}")
```

```python
a = [10, 6]
distance = math.hypot(a[0], a[1])
print(f"Euclidean distance: {distance}")
```

```python
perpendicular = 10
base = 5
hypotenuse = math.hypot(perpendicular, base)
print(f"Hypotenuse: {hypotenuse}")
```
