```python
import csv

# Define the header and data
header = ["emp_id", "emp_name", "designation", "salary"]
rows = [
    [101, "Aadhi", "Manager", 100000],
    [102, "Sam", "CEO", 200000],
    [103, "Harshitha", "Director", 100000],
    [104, "Sita", "Assistant", 50000],
    [105, "Ravi", "Executive", 75000]
]

# Write to CSV file
with open('employees.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerows(rows)

print("CSV file created successfully.")
```

```python
new_records = [
    [106, "Anil", "Analyst", 60000],
    [107, "Geetha", "Clerk", 40000],
    [108, "Kiran", "Sales", 65000]
]

# Append to CSV file
with open('employees.csv', mode='a', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(new_records)

print("New records appended to CSV file successfully.")
```

```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('employees.csv')

# Find the minimum salary
min_salary = df['salary'].min()
print(f"The minimum salary is: {min_salary}")

# Find the maximum salary
max_salary = df['salary'].max()
print(f"The maximum salary is: {max_salary}")
```

```python
# Select required columns
df_selected = df[['emp_id', 'emp_name', 'salary']]

# Write to a new CSV file
df_selected.to_csv('selected_employees.csv', index=False)

print("Selected employees saved to 'selected_employees.csv'.")
```

```python
# Define the data
text_data = [
    "101,Aadhi,Manager,100000",
    "102,Sam,CEO,200000",
    "103,Harshitha,Director,100000",
    "104,Sita,Assistant,50000",
    "105,Ravi,Executive,75000",
    "106,Anil,Analyst,60000",
    "107,Geetha,Clerk,40000",
    "108,Kiran,Sales,65000",
    "109,Meera,Technician,55000",
    "110,John,HR,70000"
]

# Write to a text file
with open('employees.txt', 'w') as file:
    for line in text_data:
        file.write(line + '\n')

print("Text file created successfully.")
```

```python
# Read the text file
with open('employees.txt', 'r') as file:
    lines = file.readlines()

# Process the data
data_list = []
for line in lines:
    emp_id, emp_name, designation, salary = line.strip().split(',')
    record = {
        "Emp_ID": int(emp_id),
        "Emp_Name": emp_name,
        "Designation": designation,
        "Salary": int(salary)
    }
    data_list.append(record)

print(data_list)
```

```python
salaries = [record['Salary'] for record in data_list]

max_salary = max(salaries)
min_salary = min(salaries)

print(f"The maximum salary is: {max_salary}")
print(f"The minimum salary is: {min_salary}")
```
