```python
lst = [1,2]
dict = {i:lst for i in lst}
print(dict)
```

```python
lst = {"a":10, "b":20, "c":30}
dict = {i:lst for i in lst}
print(dict)
```

```python
a = {"a":10, "b":20, "c":30}
dict = {a[i]:i for i in a}
print(dict)
```

```python
a = {"a":10, "b":20, "c":30}
dict = {a[i]+100:i for i in a}
print(dict)
```

```python
a = {"a":10, "b":20, "c":30}
dict = {"Letter "+i: a[i]+1000 for i in a}
print(dict)
```

```python
dict = {i:i**2 for i in range(1,5)}
print(dict)
```

```python
dict = [1,2,3,4,5,6,7,8,9]
dict2 = {i:i**2 for i in dict if i<4}
print(dict2)
```

```python
lst = [[1,2,3], ["Buddy", "in", "Multimise"]]
dict = {i:j for i,j in zip(lst[0], lst[1])}
print(dict)
```

```python
my_dict = {'a': 1, 'b': 2, 'c': 3, 'd': 4, 'e': 5}
new_dict = {key: value * 2 for key, value in my_dict.items()}
print(new_dict)
```
