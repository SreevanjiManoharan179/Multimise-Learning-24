```python
def createStack(capacity):
    top = -1
    array = [0] * capacity
    stack = [capacity, top, array]
    return stack

def isFull(s):
    return (s[1] == s[0] - 1)

def isEmpty(s):
    return s[1] == -1

def push(s, i):
    if isFull(s):
        return
    s[1] += 1
    t = s[1]
    s[2][t] = i

def Pop(s):
    if isEmpty(s):
        return None
    y = s[1]
    s[1] -= 1
    return s[2][y]
```

```python
import plotly.graph_objects as go

def moveDisksBetweenTwoPoles(src, dest, s, d):
    pole1TopDisk = Pop(src)
    pole2TopDisk = Pop(dest)

    if pole1TopDisk is None:
        if pole2TopDisk is not None:
            push(src, pole2TopDisk)
            moveDisk(d, s, pole2TopDisk)
    elif pole2TopDisk is None:
        if pole1TopDisk is not None:
            push(dest, pole1TopDisk)
            moveDisk(s, d, pole1TopDisk)
    elif (pole1TopDisk > pole2TopDisk):
        push(src, pole1TopDisk)
        push(src, pole2TopDisk)
        moveDisk(d, s, pole2TopDisk)
    else:
        push(dest, pole2TopDisk)
        push(dest, pole1TopDisk)
        moveDisk(s, d, pole1TopDisk)

def moveDisk(fromPeg, toPeg, disk):
    print(f"Move disk {disk} from '{fromPeg}' to '{toPeg}'")
    visualize_stacks()

def visualize_stacks():
    fig = go.Figure()

    for idx, (stack, name) in enumerate([(s, 'Source'), (a, 'Auxiliary'), (d, 'Destination')]):
        bars = [i for i in stack[2][:stack[1] + 1]]
        print(f"Stack {name}: {bars}")  # Print the contents of each stack
        if bars:
            fig.add_trace(go.Bar(
                x=[name] * len(bars),
                y=bars,
                text=bars,
                textposition='inside',
                name=name,
                marker_color='blue' if name == 'Source' else ('orange' if name == 'Auxiliary' else 'green')
            ))

    fig.update_layout(
        title="Tower of Hanoi",
        xaxis_title="Pegs",
        yaxis_title="Disk Size",
        barmode='stack'
    )
    fig.show()
    return fig
```

```python
def tohIterative(num_of_disks, src, aux, dest):
    s, d, a = 'Source', 'Destination', 'Auxiliary'

    if (num_of_disks % 2 == 0):
        temp = d
        d = a
        a = temp
    total_num_of_moves = int(pow(2, num_of_disks) - 1)

    for i in range(num_of_disks, 0, -1):
        push(src, i)

    for i in range(1, total_num_of_moves + 1):
        if (i % 3 == 1):
            moveDisksBetweenTwoPoles(src, dest, s, d)

        elif (i % 3 == 2):
            moveDisksBetweenTwoPoles(src, aux, s, a)

        elif (i % 3 == 0):
            moveDisksBetweenTwoPoles(aux, dest, a, d)
```

```python
import plotly.io as pio

capacity = int(input('Enter the number of disks:'))
s = createStack(capacity)
d = createStack(capacity)
a = createStack(capacity)
print("Initial State:")
fig = visualize_stacks() # Capture the returned figure object
tohIterative(capacity, s, a, d)
print("Final State:")
visualize_stacks()
```
