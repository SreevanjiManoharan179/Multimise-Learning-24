```python
colors = list(["Red", "Orange", "Yellow", "Green", "Blue", "Indigo", "Violet", "Black", "White", "Gray"])
```

```python
nColors = colors[4:7]
pColors = colors[4:7]
myColors = nColors + pColors # Concatenating lists
print(myColors)
```

```python
mid_index = len(colors) // 2
for index in range(len(colors)):
    if index == mid_index:
        print(f"Middle element: {colors[index]}")
```

```python
colors.extend(["Pink", "Purple"])
print(colors)
```

```python
removed_color = colors.pop(3)
print(f"Removed color: {removed_color}")
print(colors)
```

```python
eColors = colors[1::2]
print(eColors)
```

```python
colors.insert(6, "Cyan")
print(colors)
```

```python
print(colors[-5:-2])
```
