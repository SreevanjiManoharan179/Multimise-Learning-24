```python
sentences = ["a city named land of wonders",
             "in the holy galaxy of Andromeda",
             "on the eve of Diwali on Monday",
             "with over three lakh diya or earthen lamps",
             "lit up simultaneously on the banks of the Ganges river"]
stopw = ['for', 'a', 'of', 'the', 'and', 'to', 'in', 'on', 'with']
r = [w for sentence in sentences for w in sentence.split(' ') if w not in stopw]
print(r)
```
